```python
# 컬럼명 모두 보기
pd.set_option('display.max_columns', None)
```

```python
star_df = df[['addr', 's_name', 'tel', 'fax', 'gugun_name', 'defaultimage', 'lat', 'lot']].copy()
```

```python
df = pd.DataFrame(star_df['gugun_name'].value_counts())
```

```python
import folium
starmap = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles="cartodb positron")
```

```python
starmap
```

```python
import json
world_str=json.load(open("./skorea_municipalities_geo_simple.json", "r", encoding='utf-8'))
```

```python
geo_str = requests.get("https://t1.daumcdn.net/cfile/tistory/272C224C58B4BD540B").json()
```

```python
geo_str
```

```python
df.reset_index(inplace=True)
df.rename(columns={'count' : 'value'}, inplace=True)
folium.Choropleth(geo_str, 
                  data = df, 
                  columns=['gugun_name', 'value'],
                  fill_color= "PuRd",
                  key_on = 'feature.id').add_to(starmap)
```

```python
starmap
```

```python
# idx: 인덱스
# row: 시리즈 타입의 데이터를 반환
for idx, row in star_df[['s_name', 'lat', 'lot']].iterrows():
    folium.Marker([row.lat, row.lot], popup=row.s_name).add_to(starmap)
```
