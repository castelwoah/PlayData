```python
# 날짜, 호선, 역명, 구분 중복제거후 새로운 데이터프레임 제작
whatday = tmp['날짜'].unique()
station = tmp['역명'].unique()
onoff = tmp['구 분'].unique()
```

```python
from tqdm import tqdm

for x in tmp:
    whatday = x['날짜'].unique()
    station = x['역명'].unique()
    onoff = x['구 분'].unique()    

    for w in tqdm(whatday):
        for s in station:
            for of in onoff:
                indexlist = tmp[(tmp['구 분'] == of) & (tmp['날짜'] == w) & (tmp['역명'] == s)].index
                psum = tmp[(tmp['구 분'] == of) & (tmp['날짜'] == w) & (tmp['역명'] == s)]['people'].sum()
                for i in indexlist:
                    tmp.loc[i,'people'] = psum

```
