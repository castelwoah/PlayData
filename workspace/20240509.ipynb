```python
url = 'https://finance.naver.com/item/sise.naver?code=095570'
```

```python
r = requests.get(url)
r.text.strip()
bs = BS(r.text)
bs
```

```python
tmp = bs.findAll('table', class_='type2 type_tax')[0].findAll('tr')
x = tmp[10].text.strip().replace(" ", '_')
```

```python
x
```

```python
sise_tmp = []
for x in tmp:
    x = x.text.strip().replace(" ", '_')
    sise_tmp.append(x.split())
```

```python
sise_tmp.remove([])
```

```python
sise_df = {}
for x in sise_tmp:
    if len(x) == 4:
        sise_df[str(x[0])] = x[1]
        sise_df[str(x[2])] = x[3]
    else:
        if x[1] == '하락':
            x[1] = '-'
        else:
            x[1] = '+'
        sise_df[str(x[0])] = x[1]+x[2]
        sise_df[str(x[3])] = x[4]
```
