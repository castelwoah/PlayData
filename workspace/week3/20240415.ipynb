```python
def fibo(n):
    fibolist = []
    for x in range(n+1):
        try: 
            fibolist.append(fibolist[-1]+fibolist[-2])
        except:
            fibolist.append(x)

    return  fibolist[1:]
```

```python
fibo(10)
```

```python
# 재귀함수 사용, 
def fibo_recu(n):
    if n < 2: return n
    return fibo_recu(n-1) + fibo_recu(n-2)
```

```python
# 이터레이터
# return 대신 사용하는 것 yield
def fibo_generator():
    a,b = 0,1
    while True:
        yield b
        a , b = b, a + b
```

```python
fg = fibo_generator()
```

```python
# 이터레인터는 yield를 사용하여 잠시 멈춤사용
# 장점: 메모리를 효율적으로 사용할 수 있다
for _ in range(10):
    print(next(fg))
```
