```python
payload = """
'sid': '106'
'date': '2024-05-{}'
'page': {}
"""
```

```python
url = 'https://entertain.naver.com/now#sid=106&date={}&page={}'

r = requests.post(url, data=payload.format(16, 1))
bs = BS(r.text)
bs
```

```python
page_list = []
for d in range(1,17):
    for p in range(1,11):
        r = requests.post(url, data=payload.format(d, p))
        bs = BS(r.text)
        page_list.append(bs)
```

```python
len(page_list)
```

```python
page_list[0].findAll("div", class_='tit_area')[0].text.strip()
```

```python
text = []
for i in page_list:
    for x in i.findAll("div", class_='tit_area'):
        text.append(x.text.strip())

len(text)
```

```python
# nouns 명사만 추출
from konlpy.tag import Okt
okt = Okt()
okt.nouns(text[0])
```

```python
nouns_list=[]

for x in text:
    nouns_list.append(okt.nouns(x))
len(nouns_list)
```

```python
print(nouns_list[0])
```

```python
tmp = ''
for a in nouns_list:
    tmp += ','.join(a)
    nouns_list2 = tmp.split(',')
```
